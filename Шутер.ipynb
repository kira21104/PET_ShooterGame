```python
import random
import time

class ShooterGame:
    def __init__(self):
        self.player_pos = 5
        self.target_pos = random.randint(1, 10)

    def print_game_state(self):
        print("-" * 20)
        print(" " * (self.target_pos - 1) + "T" + " " * (10 - self.target_pos))
        print("-" * 20)
        print(" " * (self.player_pos - 1) + "P" + " " * (10 - self.player_pos))
        print("-" * 20)

    def shoot(self):
        print("Вы стреляете...")
        time.sleep(1)
        hit = random.choice([True, False])
        if hit and self.player_pos == self.target_pos:
            print("Попадание! Цель поражена.")
            return True
        else:
            print("Промах. Цель ушла.")
            return False

    def play(self):
        print("Добро пожаловать в игру 'Текстовый Шутер'!")
        while True:
            self.print_game_state()
            action = input("Выберите действие (a - влево, d - вправо, s - стрелять, q - выход): ")
            if action == "a" and self.player_pos > 1:
                self.player_pos -= 1
            elif action == "d" and self.player_pos < 10:
                self.player_pos += 1
            elif action == "s":
                if self.shoot():
                    print("Поздравляю! Вы победили.")
                    break
            elif action == "q":
                print("Игра завершена.")
                break
            else:
                print("Неверная команда. Попробуйте еще раз.")

if __name__ == "__main__":
    game = ShooterGame()
    game.play()
```
